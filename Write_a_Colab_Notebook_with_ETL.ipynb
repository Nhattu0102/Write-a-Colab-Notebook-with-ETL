```python
def change_data_type(df, cols, change_data_type):
  df[cols] = df[cols].astype(change_data_type)
  return df
```

```python
def lower_cols(df, cols):
  df[cols] = df[cols].str.lower()
  return df
```

```python
def fill_na(df, cols, value):
  df[cols] = df[cols].fillna(value)
  return df
```

## Enrollies' working experience

```python
enrollies_experience['experience'].unique()
```

```python
enrollies_experience.head()
```

```python
def classify_experience(experience):
    if experience == '>20':
        return '>20'
    elif experience == '<1':
        return '<1'
    try:
        experience = int(experience)
        if 1 <= experience <= 5:
            return '1-5'
        elif 6 <= experience <= 10:
            return '6-10'
        elif 11 <= experience <= 15:
            return '11-15'
        elif 16 <= experience <= 20:
            return '16-20'
        else:
            return '>20'
    except (ValueError,TypeError):
        return np.nan
```

```python
enrollies_experience['experience_classified'] = enrollies_experience['experience'].apply(classify_experience)
```
